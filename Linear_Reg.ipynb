```python
# Importing requisite librariies
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"D:\Datasets\Salary_Data.csv")
# Overview of the dataset
print(f"\nHere is the overview of dataset \n{df.head()}\n")
print(f"Number of rows and columns in the data set are: {df.shape}")
```

```python
# Plotting the dataset for overlook
plt.title("Salary increase with experience", color = "Red")
plt.xlabel("No of years", color = "#2C3E50")
plt.ylabel("Salary increase", color = "#000000")
plt.scatter(df.YearsExperience,df.Salary)
plt.show()
```

```python
# Training the model
lnr_model = LinearRegression()
training_model = lnr_model.fit(df[["YearsExperience"]],df.Salary)
print(training_model)
```

```python
#Making prediction for salary
predict_value = lnr_model.predict([[1.9]])
print(predict_value)
```

```python
model_acc = lnr_model.score(df[["YearsExperience"]],df.Salary)
print(f"Accuracy of the Linear Model is {model_acc * 100:.2f} %")
```

```python
# calculating the coefficient
coefficient = lnr_model.coef_
print(coefficient)
```

```python
# calculating the coefficient
intercept = lnr_model.intercept_
print(intercept)
```

```python
y_predicted = coefficient *  1.9 + intercept
print(f"Value calculation using slope-intercept formula is : {y_predicted}")
print(f"Value calculated by the model is: {predict_value}\nThe difference between values: {y_predicted-predict_value}")
```

```python
plt.scatter(df["YearsExperience"], df["Salary"])
plt.xlabel("No of years", color="#2C3E50")
plt.ylabel("Salary increase", color="#000000")

plt.plot(df["YearsExperience"], lnr_model.predict(df[["YearsExperience"]]))

plt.show()
```
